#1) Quyida berilgan datasetni yuklab olib, ustun nomlarini koʻring va ularni maʼno jihatdan tushunarli shaklga keltiring. `Unnamed: 0` ustunini tashlab yuboring!

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/restaurant_sales.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.columns = [col.lower().replace(' ', '_') for col in df.columns]

print("Yangi ustun nomlari:", df.columns.tolist())
df.head()
```

#3) Har bir ustundagi boʻsh qiymatlarni aniqlang va ularni toʻldirish strategiyasini ishlab chiqing.

```python
categorical_cols = ['category', 'item', 'payment_method']
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

df['price'] = df['price'].fillna(df['price'].median())
df['quantity'] = df['quantity'].fillna(df['quantity'].median())

df['order_total'] = df['order_total'].fillna(df['price'] * df['quantity'])

df.dropna(subset=['order_date'], inplace=True)

print("Tozalashdan keyingi bo'sh qiymatlar:")
print(df.isnull().sum())
```

#4) Har bir ustundagi takroriy qiymatlarni toping va ularga qanday oʻzgartirish kiritish kerakligini hal qiling.

```python
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df.columns = [col.lower().replace(' ', '_') for col in df.columns]

text_cols = ['category', 'item', 'payment_method']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

df['category'] = df['category'].replace('nan', 'unknown')
df['item'] = df['item'].replace('nan', 'unknown')
df['price'] = df['price'].fillna(df['price'].median())
df['quantity'] = df['quantity'].fillna(df['quantity'].median())

df['order_total'] = df['order_total'].fillna(df['price'] * df['quantity'])

df.drop_duplicates(inplace=True)

df['order_date'] = pd.to_datetime(df['order_date'])

print("Dataset hajmi:", df.shape)
print("\nBo'sh qiymatlar soni:\n", df.isnull().sum())
df.head()
```

```python
numeric_cols = ['price', 'quantity', 'order_total']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['price'] = df['price'].fillna(df['price'].median())
df['quantity'] = df['quantity'].fillna(df['quantity'].median())
df['order_total'] = df['price'] * df['quantity']
print("Ustunlarning yakuniy turlari:")
print(df[numeric_cols].dtypes)

df.head()
```

#7) Maʼlumotlar turini oʻzgartirish.
 * Ustunlarning maʼlumot turlarini tekshiring. Notoʻgʻri turlarga ega boʻlgan ustunlarni toʻgʻri formatga oʻtkazing.

```python
print("O'zgartirishdan oldingi turlar:\n", df.dtypes)

df['order_date'] = pd.to_datetime(df['order_date'])

df['quantity'] = df['quantity'].astype(int)
df['price'] = df['price'].astype(float)
df['order_total'] = df['order_total'].astype(float)
df['category'] = df['category'].astype('category')
df['payment_method'] = df['payment_method'].astype('category')

print("\nO'zgartirishdan keyingi turlar:\n", df.dtypes)
df.head()
```

#8) Mahsulot nomlarida noto‘g‘ri yoki nomaʼlum belgilar (?, NaN, #) mavjudligini tekshiring va ularni tuzating.

```python
import re

def clean_text(text):
    if pd.isna(text) or str(text).lower() == 'nan':
        return "unknown"
    text = re.sub(r'[^a-zA-Z0-9\s]', '', str(text))

    text = " ".join(text.split())

    return text.lower()

df['item'] = df['item'].apply(clean_text)
df['category'] = df['category'].apply(clean_text)
print("Tozalashdan keyingi 'unknown' qiymatlar soni:")
print(df[df['item'] == 'unknown']['item'].count())

df.head(10)
```

#9) Manfiy narx yoki 0 ga teng narx qiymatlarini toping va ularni qanday oʻzgartirish kerakligini hal qiling.

```python
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
def clean_text(text):
    if pd.isna(text) or str(text).lower() == 'nan':
        return "unknown"
    text = re.sub(r'[^a-zA-Z0-9\s]', '', str(text))
    return " ".join(text.split()).lower()

text_cols = ['category', 'item', 'payment_method']
for col in text_cols:
    df[col] = df[col].apply(clean_text)
numeric_cols = ['price', 'quantity', 'order_total']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.loc[df['price'] <= 0, 'price'] = np.nan
df['price'] = df['price'].fillna(df.groupby('category')['price'].transform('median'))
df['price'] = df['price'].fillna(df['price'].median())

df['quantity'] = df['quantity'].fillna(df['quantity'].median())

df['order_total'] = df['price'] * df['quantity']

df['order_date'] = pd.to_datetime(df['order_date'])
df['quantity'] = df['quantity'].astype(int)
df['price'] = df['price'].astype(float)
df['category'] = df['category'].astype('category')

df.drop_duplicates(inplace=True)

print("--- Dataset haqida umumiy ma'lumot ---")
df.info()

print("\n--- Bo'sh qiymatlar soni ---")
print(df.isnull().sum())

print("\n--- Statistik ko'rsatkichlar ---")
print(df.describe())

df.head()
```

#10) Boʻsh toifalar (kategoriya) mavjudligini tekshiring va toʻgʻri kategoriya berish strategiyasini ishlab chiqing.

```python
df['category'] = df['category'].replace(['unknown', 'nan', 'none', ''], np.nan)
df['category'] = df['category'].fillna(df.groupby('item')['category'].transform('first'))
def fill_by_keyword(row):
    if pd.isna(row['category']):
        item = str(row['item']).lower()
        if any(x in item for x in ['tea', 'coffee', 'juice', 'cola', 'water']):
            return 'beverages'
        elif any(x in item for x in ['cake', 'ice cream', 'pudding']):
            return 'desserts'
        elif any(x in item for x in ['pizza', 'burger', 'steak', 'pasta']):
            return 'main course'
    return row['category']

df['category'] = df.apply(fill_by_keyword, axis=1)
df['category'] = df['category'].fillna('others')

print("Bo'sh kategoriyalar qoldimi:", df['category'].isnull().sum())
print("\nYangilangan kategoriyalar ro'yxati:\n", df['category'].value_counts())
```

#11) Bir xil, lekin turlicha yozilgan qiymatlar (Coffee, cofee, COFEE) mavjudligini tekshiring va tuzating.

```python
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df.columns = [col.lower().replace(' ', '_') for col in df.columns]

def clean_text(text):
    if pd.isna(text) or str(text).lower() == 'nan':
        return "unknown"
    text = re.sub(r'[^a-zA-Z0-9\s]', '', str(text))
    text = " ".join(text.split()).lower()
    return text

text_cols = ['category', 'item', 'payment_method']
for col in text_cols:
    df[col] = df[col].apply(clean_text)
corrections = {
    'cofee': 'coffee', 'cofeee': 'coffee', 'burgerr': 'burger',
    'piza': 'pizza', 'steyk': 'steak', 'teaa': 'tea'
}
df['item'] = df['item'].replace(corrections)
numeric_cols = ['price', 'quantity', 'order_total']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.loc[df['price'] <= 0, 'price'] = np.nan
df['price'] = df['price'].fillna(df.groupby('category')['price'].transform('median'))
df['price'] = df['price'].fillna(df['price'].median())
df['quantity'] = df['quantity'].fillna(df['quantity'].median())

item_cat_map = df.dropna(subset=['category']).set_index('item')['category'].to_dict()
df['category'] = df['category'].replace('unknown', np.nan)
df['category'] = df['category'].fillna(df['item'].map(item_cat_map))
df['category'] = df['category'].fillna('others')

df['order_total'] = df['price'] * df['quantity']

df['order_date'] = pd.to_datetime(df['order_date'])
df['quantity'] = df['quantity'].astype(int)
df['category'] = df['category'].astype('category')

df.drop_duplicates(inplace=True)

print("--- Dataset haqida ma'lumot ---")
df.info()
print("\n--- Statistik ko'rsatkichlar ---")
print(df.describe())
print("\n--- Bo'sh qiymatlar ---")
print(df.isnull().sum())

df.head()
```

#12) `Customer ID` yoki boshqa identifikatorlarning takrorlanishini tekshiring.

```python
customer_revisits = df['customer_id'].value_counts()
loyal_customers = customer_revisits[customer_revisits > 1]

print(f"Jami noyob mijozlar soni: {df['customer_id'].nunique()}")
print(f"Qayta tashrif buyurgan (sodiq) mijozlar soni: {len(loyal_customers)}")
order_items_count = df['order_id'].value_counts()
print(f"Bitta buyurtmada eng ko'p mahsulot sotib olingan soni: {order_items_count.max()}")

duplicate_mask = df.duplicated(subset=['order_id', 'customer_id', 'item', 'order_date'], keep='first')
duplicate_count = duplicate_mask.sum()

if duplicate_count > 0:
    print(f"\nAniqlangan texnik xato dublikatlar soni: {duplicate_count}")
    df = df.drop_duplicates(subset=['order_id', 'customer_id', 'item', 'order_date'], keep='first')
    print("Texnik dublikatlar muvaffaqiyatli o'chirildi.")
else:
    print("\nTexnik xato dublikatlar aniqlanmadi.")
print(f"\nIDsi yo'q mijozlar soni: {df['customer_id'].isnull().sum()}")
print(f"IDsi yo'q buyurtmalar soni: {df['order_id'].isnull().sum()}")
print("\nEng ko'p daromad keltirgan 5 ta mijoz (Top Customers):")
top_customers = df.groupby('customer_id')['order_total'].sum().sort_values(ascending=False)
print(top_customers.head())
```

#13) Mahsulotning ortiqcha taʼriflarini olib tashlang (Water (0.5), Water (1.0), Water → faqat Water qoldirish).

```python
def clean_item_description(item_name):
    if pd.isna(item_name):
        return "unknown"
    cleaned = re.sub(r'\(.*?\)', '', str(item_name))

    cleaned = cleaned.split('—')[0].split('-')[0]
    return cleaned.strip().lower()
df['item'] = df['item'].apply(clean_item_description)

print("Tozalashdan keyingi noyob mahsulotlar (masalan, barcha Water'lar birlashdi):")
print(df['item'].unique()[:15])
print(f"\nBirlashishdan keyingi umumiy qatorlar soni: {len(df)}")
```

#16) Mijozlarning qaysi kunlarda koʻproq xarid qilgani bo‘yicha tahlil qiling.

```python
df['day_of_week'] = df['order_date'].dt.day_name()

weekly_sales = df['day_of_week'].value_counts()
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sales = weekly_sales.reindex(days_order)

print("--- Haftaning kunlari bo'yicha buyurtmalar soni ---")
print(weekly_sales)

weekly_revenue = df.groupby('day_of_week')['order_total'].sum().reindex(days_order)
print("\n--- Haftaning kunlari bo'yicha umumiy tushum ---")
print(weekly_revenue)
```

#17) Narxi oʻrtacha narxdan ikki barobar yuqori boʻlgan mahsulotlarni toping.

```python
average_price = df['price'].mean()
print(f"Barcha mahsulotlarning o'rtacha narxi: {average_price:.2f}")
threshold = average_price * 2
print(f"2 barobar yuqori bo'lgan chegara (Threshold): {threshold:.2f}")

expensive_items = df[df['price'] > threshold][['item', 'category', 'price']].drop_duplicates()

print("\n--- O'rtacha narxdan 2 barobar qimmat mahsulotlar ---")
print(expensive_items.sort_values(by='price', ascending=False))
```

#18) Mahsulotlar kategoriyalari bo‘yicha tahlil qiling va eng ko‘p sotiladigan kategoriyalarni aniqlang.

```python
category_analysis = df.groupby('category').agg({
    'quantity': 'sum',
    'order_total': 'sum'
}).sort_values(by='quantity', ascending=False)

total_quantity = category_analysis['quantity'].sum()
category_analysis['market_share_%'] = (category_analysis['quantity'] / total_quantity) * 100

print("--- Kategoriyalar bo'yicha sotuv tahlili ---")
print(category_analysis)

top_category = category_analysis.index[0]
print(f"\nEng ommabop kategoriya: {top_category.upper()}")
```

#19) Har bir mijozning oʻrtacha xarid miqdorini (`groupby("customer_id")`) hisoblang.

```python
customer_avg_stats = df.groupby('customer_id').agg({
    'order_total': 'mean',
    'quantity': 'mean'
}).rename(columns={
    'order_total': 'avg_spent',
    'quantity': 'avg_quantity'
})
print("--- Har bir mijozning o'rtacha xarid ko'rsatkichlari (Top 10) ---")
print(customer_avg_stats.sort_values(by='avg_spent', ascending=False).head(10))

overall_avg_quantity = customer_avg_stats['avg_quantity'].mean()
print(f"\nRestoran bo'yicha bitta mijozning o'rtacha xarid miqdori: {overall_avg_quantity:.2f} ta mahsulot")
```

#20) Qanday omillar sotuv hajmiga ta’sir qiladi? Bu bo‘yicha umumiy xulosa chiqaring. Javobni quyida kommentariya sifatida yozing!

```python
import pandas as pd
import numpy as np
import re

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/restaurant_sales.csv"
df = pd.read_csv(url)
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')
df.columns = df.columns.str.lower().str.replace(' ', '_')

def clean_item_name(text):
    if pd.isna(text): return "unknown"
    text = re.sub(r'\(.*?\)', '', str(text))
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return " ".join(text.split()).lower()

for col in ['category', 'item', 'payment_method']:
    df[col] = df[col].apply(clean_text)

df['item'] = df['item'].replace({'cofee': 'coffee', 'piza': 'pizza', 'burgerr': 'burger'})

cols_to_fix = ['price', 'quantity', 'order_total']
df[cols_to_fix] = df[cols_to_fix].apply(pd.to_numeric, errors='coerce')

df.loc[df['price'] <= 0, 'price'] = np.nan
df['price'] = df['price'].fillna(df.groupby('category')['price'].transform('median'))
df['quantity'] = df['quantity'].fillna(df['quantity'].median())
df['order_total'] = df['price'] * df['quantity']
df['order_date'] = pd.to_datetime(df['order_date'])
df['category'] = df['category'].replace('unknown', np.nan)
df['category'] = df['category'].fillna(df.groupby('item')['category'].transform('first')).fillna('others')
df.drop_duplicates(inplace=True)
top_5 = df.groupby('item')['quantity'].sum().nlargest(5)
customer_avg = df.groupby('customer_id')['order_total'].mean().mean()
df['day'] = df['order_date'].dt.day_name()
top_day = df['day'].value_counts().idxmax()
premium_list = df[df['price'] > (df['price'].mean() * 2)]['item'].unique()
top_cat = df.groupby('category')['quantity'].sum().idxmax()
print(f"1. Top 5 mahsulot:\n{top_5}")
print(f"\n2. Mijozlarning o'rtacha xarajat summasi: {customer_avg:.2f}")
print(f"3. Savdo eng qizigan kun: {top_day}")
print(f"4. Eng ommabop kategoriya: {top_cat}")
print(f"5. Premium mahsulotlar soni: {len(premium_list)}")
```
